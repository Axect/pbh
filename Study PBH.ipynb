## Basic Scalar Field Inflation
*****
### Scalar Field Action

\begin{equation}
S = \int d^4x\sqrt{-g}\left[\frac{1}{2}R + \frac{1}{2}g^{\mu\nu}\partial_\mu\phi\partial_\nu\phi - V(\phi)\right] = S_{EH} + S_{\phi}
\end{equation}

### Energy-Momentum Tensor
\begin{equation}
T_{\mu\nu}^{(\phi)} \equiv -\frac{2}{\sqrt{-g}}\frac{\delta S_\phi}{\delta g^{\mu\nu}} = \partial_\mu\phi\partial_\nu\phi - g_{\mu\nu}\left(\frac{1}{2}\partial^\sigma\phi\partial_\sigma\phi + V(\phi)\right)
\end{equation}

* Consider FRW metric :

\begin{align}
\rho_\phi &= \frac{1}{2}\dot{\phi}^2 + V(\phi) \\
p_\phi &= \frac{1}{2}\dot{\phi}^2 -V(\phi)
\end{align}

* The eq of State :

\begin{equation}
w_\phi \equiv \frac{p_\phi}{\rho_{\phi}} = \frac{\frac{1}{2}\dot{\phi}^2 - V}{\frac{1}{2}\dot{\phi}^2 + V}
\end{equation}

> * Negative pressure : $w_\phi < 0$
> * Accelerated Expansion : $w_\phi < -\frac{1}{3}$

### Slow-Roll Inflation

\begin{equation}
\frac{\ddot{a}}{a} = -\frac{1}{6}(\rho_\phi + 3p_\phi) = H^2 (1-\epsilon)
\end{equation}

* Hubble Slow-Roll Parameter

\begin{equation}
\epsilon = -\frac{\dot{H}}{H^2} = -\frac{d \ln H}{dN} \quad (dN=Hdt)
\end{equation}

* Hubble Second Slow-Roll Parameter

\begin{equation}
\eta = -\frac{\ddot{\phi}}{H\dot{\phi}} = \epsilon - \frac{1}{2\epsilon}\frac{d\epsilon}{dN}
\end{equation}

* The Slow-Roll Conditions : $$\epsilon, |\eta|\ll 1$$

* Potential Slow-Roll Parameters

\begin{align}
\epsilon_v(\phi) &\equiv \frac{M_{pl}^2}{2}\left(\frac{V_{,\phi}}{V}\right)^2 \\
\eta_v(\phi) &\equiv M_{pl}^2 \frac{V_{,\phi\phi}}{V}
\end{align}

* The Slow-Roll Conditions Revisted : $$\epsilon_v, |\eta_v| \ll 1$$

* Relation : $$ \epsilon \approx \epsilon_v,\quad \eta \approx \eta_v - \epsilon_v$$

* The number of $e$-folds

\begin{align}
 N(\phi) &\equiv \ln \frac{a_{end}}{a} \\
         &= \int H dt = \int_\phi^{\phi_{end}}\frac{H}{\dot{\phi}}d\phi \approx \int_{\phi_{end}}^\phi \frac{V}{V_{,\phi}} d\phi \\
         &= \int_{\phi_{end}}^\phi \frac{d\phi}{\sqrt{2\epsilon}} \approx \int_{\phi_{end}}^\phi \frac{d\phi}{\sqrt{2\epsilon_v}}
\end{align}

* To Solve the Horizon & Flatness : $$N_{tot} \geq 60$$

### Cosmological Perturbations

* Power Spectrum : $$ \Delta_s^2 \equiv \Delta_{\mathcal{R}}^2 = \frac{k^3}{2\pi^2}P_{\mathcal{R}}(k)$$

* Scalar Spectral index : $$\eta_s - 1 \equiv \frac{d\ln\Delta_s^2}{d\ln k}$$

* Running Spectral index : $$\alpha_s \equiv \frac{dn_s}{d\ln k}$$

* Power law approx : 
$$ \Delta_s^2(k) =  A_s(k_\star) \left(\frac{k}{k_\star} \right)^{n_s(k_\star)-1+\frac{1}{2}\alpha_s(k_\star)\ln(k/k_\star)}$$
($k_\star$ is an arbitrary reference or pivot scale)

* Tensor-to-Scalar Ratio
$$ r \equiv \frac{\Delta_t^2(k)}{\Delta_s^2(k)} $$

### Slow Roll Results

\begin{equation}
\Delta_s^2(k) \approx \frac{V}{24\pi^2M_p^4\epsilon_v}\Big|_{k=aH},\quad \Delta_t^2(k) \approx \frac{2V}{3\pi^2 M_p^4}\Big|_{k=aH}
\end{equation}

\begin{align}
 n_s - 1 &\equiv \frac{d\ln\Delta_s^2}{d\ln k} = 2\eta_v - 6\epsilon_v \\
 n_t &\equiv \frac{d\ln\Delta_t^2}{d\ln k} = -2\epsilon_v \\
 r & \equiv \frac{\Delta_t^2}{\Delta_s^2} = 16\epsilon_v \\
 \frac{\Delta\phi}{M_p} &\approx \left(\frac{r}{0.01} \right)^{1/2} \quad \text{(The Lyth Bound)}
\end{align}

## Higgs Inflation
*****
### Action

\begin{equation}
 S = \int d^4x\sqrt{-g}\left[\frac{M_p^2}{2}R + \frac{1}{2}\xi\phi^2 R - \frac{1}{2}g^{\mu\nu}\partial_\mu\phi\partial_\nu\phi - \frac{1}{4}\lambda\phi^4 \right]
\end{equation}

### Einstein Frame 
($g_{\mu\nu} \rightarrow g^E_{\mu\nu} = \Omega^2g_{\mu\nu}, \quad \Omega^2 = 1+\xi\phi^2/M_p^2$)

* Action

\begin{equation}
S^E = \int d^4x\sqrt{-g^E}\left[\frac{M_p^2}{2}R^E - \frac{1}{2}g^{E\mu\nu}\partial_\mu\varphi\partial_\nu\varphi - V \right]
\end{equation}

\begin{equation}
\left(\frac{d\varphi}{d\phi} \right)^2 = \frac{1+(1+6\xi)\xi\phi^2/M_p^2}{(1+6\xi\phi^2/M_p^2)^2}
\end{equation}

* Potential
$$V = \frac{\lambda\phi^4}{4\left(1+\xi\frac{\phi^2}{M_p^2}\right)^2}\,\,\rightarrow\,\,\frac{\lambda M_p^4}{4\xi^2} \quad \text{(At High field value)}$$

* Potential Slow-Roll Parameters

\begin{align}
\epsilon_v(\phi) &\equiv \frac{M_{pl}^2}{2}\left(\frac{V_{,\varphi}}{V}\right)^2 \\
\eta_v(\phi) &\equiv M_{pl}^2 \frac{V_{,\varphi\varphi}}{V}
\end{align}

> * Explicit Formula :

> ($\lambda' = M_p \frac{d\lambda}{d\phi}, \xi' = M_p\frac{d\xi}{d\phi}$)
\begin{equation}
 \epsilon_v(\phi) =  \frac{8M_p^2}{\phi^2} \frac{\left[1 + \frac{\lambda'}{4\lambda}\left( 1+\xi \left(\frac{\phi}{M_p}\right)^2\right) \frac{\phi}{M_p}   -\frac{\xi'}{2}\left(\frac{\phi}{M_p} \right)^3 \right]}{1+(1+6\xi)\xi\left(\frac{\phi}{M_p^2} \right)}
\end{equation}

### Super Slow Roll (SSR) 
* Conditions :  
($\epsilon_v = 0$)

\begin{equation}
1+\frac{\lambda'}{4\lambda}\left(1+\xi\frac{\phi^2}{M_p^2}\right)\frac{\phi}{M_p}-\frac{1}{2}\xi'\frac{\phi^3}{M_p^3} = 0
\end{equation}

> * Substitute to Cubic Equation  
>$\left(\dfrac{\phi}{M_p}\Big|_{SSR} \approx x_*,\quad \alpha = \dfrac{\lambda'}{4\lambda},\quad \beta=\dfrac{\lambda'}{4\lambda}\xi - \dfrac{\xi'}{2} \right)$

>\begin{equation}
 1 + \alpha x_* + \beta x_*^3 = 0
\end{equation}

\begin{align}
N(\varphi) &= \int_\varphi^{\varphi_0}\frac{d\varphi}{\sqrt{2\epsilon_v}} \quad \text{($\phi \rightarrow \frac{\phi}{M_p}$)}\\
